* Problema 2: $y - 5 = 12$Paso 1 (Identificar): El número 5 está restando a la $y$.Paso 2 (Despejar): Lo pasamos al lado derecho con la operación opuesta, que es la suma.$$y = 12 + 5$$Resultado: $y = 17$

```python
# Pedimos el resultado de la igualdad y el valor que resta
resultado = int(input("Introduce el resultado : "))
resta = int(input("Introduce el valor que resta : "))

# Despeje: y = resultado + resta
y = resultado + resta

print(f"El valor de y es: {y}")
```

Problema 4: $12 = m + 3m$Paso 1 (Agrupar): En el lado derecho, sumamos las $m$. Recuerda que $m$ es lo mismo que $1m$.$$1m + 3m = 4m$$$$12 = 4m$$Paso 2 (Despejar): El 4 está multiplicando a la $m$. Pasa al otro lado dividiendo.$$m = \frac{12}{4}$$Resultado: $m = 3$

```python
total = int(input("Introduce el total : "))
coeficiente1 = int(input("Introduce el primer coeficiente de m : "))
coeficiente2 = int(input("Introduce el segundo coeficiente de m : "))

# Sumamos las m y dividimos: m = total / (coef1 + coef2)
m = total / (coeficiente1 + coeficiente2)

print(f"El valor de m es: {int(m)}")
```

Problema 6: $5b + 2b - 3b = 16$Paso 1 (Agrupar): Resolvemos las operaciones de los términos con $b$ en el lado izquierdo.$$(5 + 2 - 3)b = 16$$$$4b = 16$$Paso 2 (Despejar): El 4 que multiplica a $b$ pasa al otro lado dividiendo.$$b = \frac{16}{4}$$Resultado: $b = 4$

```python
total = int(input("Introduce el total : "))
# Operamos los coeficientes: 5 + 2 - 3
suma_coeficientes = 5 + 2 - 3

# Despeje: b = total / suma_coeficientes
b = total / suma_coeficientes

print(f"El valor de b es: {int(b)}")
```

Problema 8: $7y - 5 = 3y + 7$Paso 1 (Organizar): Pasamos las "y" al lado izquierdo y los números al derecho. El $3y$ pasa restando y el $5$ pasa sumando.$$7y - 3y = 7 + 5$$Paso 2 (Simplificar): Restamos las variables y sumamos los números.$$4y = 12$$Paso 3 (Despejar): El 4 pasa dividiendo.$$y = \frac{12}{4}$$Resultado: $y = 3$

```python
# Datos: 7y - 5 = 3y + 7
coef_izq = int(input("Coeficiente de y a la izquierda : "))
num_izq = int(input("Número a la izquierda : "))
coef_der = int(input("Coeficiente de y a la derecha : "))
num_der = int(input("Número a la derecha : "))

# Agrupamos: (7 - 3)y = 7 + 5
y = (num_der - num_izq) / (coef_izq - coef_der)

print(f"El valor de y es: {int(y)}")
```

Problema 10: $8n - 6 = 4n + 10$Paso 1 (Organizar): Movemos $4n$ a la izquierda (restando) y el $6$ a la derecha (sumando).$$8n - 4n = 10 + 6$$Paso 2 (Simplificar): Resolvemos ambos lados.$$4n = 16$$Paso 3 (Despejar): El 4 pasa dividiendo.$$n = \frac{16}{4}$$Resultado: $n = 4$

```python
# Datos: 8n - 6 = 4n + 10
n_izq = int(input("Coeficiente de n a la izquierda : "))
resta_izq = int(input("Número restando a la izquierda : "))
n_der = int(input("Coeficiente de n a la derecha : "))
suma_der = int(input("Número sumando a la derecha : "))

# Despeje: (8 - 4)n = 10 + 6
n = (suma_der + resta_izq) / (n_izq - n_der)

print(f"El valor de n es: {int(n)}")
```

Problema 12: $10b + 5 = 2b + 37$Paso 1 (Organizar): Pasamos $2b$ a la izquierda restando y el $5$ a la derecha restando.$$10b - 2b = 37 - 5$$Paso 2 (Simplificar): Resolvemos las operaciones.$$8b = 32$$Paso 3 (Despejar): El 8 pasa dividiendo.$$b = \frac{32}{8}$$Resultado: $b = 4$

```python
# Datos: 10b + 5 = 2b + 37
b_izq = int(input("Coeficiente de b a la izquierda : "))
num_izq = int(input("Número sumando a la izquierda : "))
b_der = int(input("Coeficiente de b a la derecha : "))
num_der = int(input("Número a la derecha : "))

# Despeje: (10 - 2)b = 37 - 5
b = (num_der - num_izq) / (b_izq - b_der)

print(f"El valor de b es: {int(b)}")
```
